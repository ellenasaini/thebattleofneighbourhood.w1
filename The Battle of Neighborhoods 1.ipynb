## Opening a Malay Cuisine Restaurant in Kuala Lumpur

# 1. Introduction
## background of Malaysia and target audience
Kuala Lumpur is a robust city in Southeast Asia that proudly claims they have the most scrumtious food to ever exist.

Besides that, Petronas Twin Towers, pair of the tallest skyscraper office buildings are located in Kuala Lumpur.

The city-state is strongly populated and home to nearly 1.8 million people of four ethnic groups: 
The Chinese, Malay and Indian majority.


The multi-culture has contributed to the diverse culinary field. Though 
One can easily spot an eclectic mix of restaurants from all over the world.

It is still highly profitable to open a restaurant on the main streets and malls of Kuala Lumpur. 

Taking into account the high population densities and the increasing number of tourists. 

Since Malays occupy most of the inhabitants,the restaurant is Malay.

This capstone project we will help potential stakeholders select prime locations to open
the restaurant.


## Target Audiences

- Group 1 : Potential stakeholders who want to invest in opening a Malay restaurant. This project will help them narrow down their location options.

- Group 2: Citizens who want to have a clearer idea of KL neighbourhoods . This project is helpful  when they want to find the top popular venues around a specific area.

- Group 3: Tourists in KL. This project provides them with an idea of every neighbourhood and the list of restaurants, and would be helpful when they want to find a restaurant for meals but have no idea where to go.

## Creating and Solving Problems

We believe a stakeholder would like to open a Malay restaurant in Kuala Lumpur, and have not yet agreed on the spot.
He is not well acquainted with the city and has asked us to suggest areas or locations where he can open his restaurant.
When choosing the optimal spot, other factors must be taken into account.

- choosing a location where Malay food is not common among the neighbouorhood.
- areas surrounded by entertainment or work places.
