# Data Collection

## In this project we will explore the neighbourhoods in the level of Planning Area. Retrieving venue data around each Planning Area and select Planning Areas as the 'best' locations to open a Malay restaurant.





## https://www.dbkl.gov.my/pskl2020/english/strategic_zone/index.htm  is the link for Data of Kuala Lumpur Planning Area boundaries


# I used Foursquare's 'explore' API call to get the information of venues around each Planning Area. Foursquare is a location data provider with information about all manner of venues and events within an area of interest. Such information includes venue names, locations, latitudes, longitudes, etc. For each Planning Area, we have chosen the radius to be 2000 meters. We only captured the venue data useful for us and populated it into a dataframe named venues_df. Figure 3 describes the detailed venue data in the dataframe venues_df.
